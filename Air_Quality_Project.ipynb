# Model: Random Forest Regressor

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
x1 = air1[['City', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]
y1 = air1[['AQI']]
```

```python
y1 = np.ravel(y1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y1,train_size = 0.8, random_state = 3)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train,y_train)
print(regr.predict(X_test))
```

```python
pred_rf = regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pred_rf)
```

# Model: Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
print(reg.predict(X_test))
```

```python
pred_lr = reg.predict(X_test)
```

```python
r2_score(y_test, pred_lr)
```

```python
import numpy as np
from flask import Flask, request, jsonify, render_template
from flask_ngrok import run_with_ngrok
import pickle


app = Flask(__name__)
model = pickle.load(open('/content/drive/My Drive/Aqi0.pkl','rb'))
run_with_ngrok(app)

@app.route('/')
def home():
  return render_template("index.html")

@app.route('/predict',methods=['GET'])
def predict():


    '''
    For rendering results on HTML GUI
    '''
    city = float(request.args.get('city'))
    pm25 = float(request.args.get('pm25'))
    pm10 = float(request.args.get('pm10'))
    no = float(request.args.get('no'))
    no2 = float(request.args.get('no2'))
    nox = float(request.args.get('nox'))
    nh3 = float(request.args.get('nh3'))
    co = float(request.args.get('co'))
    so3 = float(request.args.get('so3'))
    o3 = float(request.args.get('o3'))
    benzene = float(request.args.get('benzene'))
    toluene = float(request.args.get('toluene'))
    xylene = float(request.args.get('xylene'))





    prediction = model.predict([[city, pm25, pm10, no, no2, nox, nh3, co, so3, o3, benzene, toluene, xylene]])


    return render_template('index.html', prediction_text='AQI= : {}'.format(prediction))


app.run()
```
